```
from bs4 import BeautifulSoup
import requests
```

```
def get_episodes_links():
  """Return the links to the scripts of all Seinfeld episodes"""
  episodes_page = 'http://www.seinfeldscripts.com/seinfeld-scripts.html'
  r = requests.get(episodes_page)
  page_content = BeautifulSoup(r.content, "html.parser")
  # get a list of 'href' of all 'a' tags 
  links = list(map(lambda x: x['href'].strip(), page_content.find_all('a', href=True)))

  episodes_links = []
  flag = False
  base_url = 'http://www.seinfeldscripts.com/'
  for episode in links:
    # first episode
    if 'Chronicles' in episode:
      flag = True

    # episodes with highlights, no scripts
    highlight_episodes = ['Highlights', 'The-Clip-Show']
    if any(hl in episode for hl in highlight_episodes):
      continue

    if flag:
      episodes_links.append(base_url + episode)

    # last episode
    if 'Finale' in episode:
      break
    
  return episodes_links 

for episode in get_episodes_links():
  print(episode)
```
